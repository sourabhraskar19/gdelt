```python
def display_news(bbox, pop_d, argOpacity=0.6):
        centerx = (bbox[0] + bbox[2]) * 1.0 / 2
        centery = (bbox[1] + bbox[3]) * 1.0 / 2
        

        m = folium.Map(location=[centery, centerx], zoom_start=2) 
        for row in pop_d.itertuples():
            #c=geoAPI()
            #if not row.sentiment_summary=="absent":
                #pos=c.search(row.NAME_2)['result']['results'][0]['geometry']['location']
            popuplocation=[row.lat,row.lon]
            m.add_child(folium.Marker(location=popuplocation,popup=row.title +','+ row.provider))
            #else:
            #    pass

        #m.add_child(m)
        folium.LayerControl().add_to(m)

        return m
```
